```python
1. # Meminta input suhu dari pengguna
suhu = float(input("Masukkan suhu air pendingin (°C): "))

# Percabangan kondisi suhu
if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")
```

Penjelasan:

float(input(...)) digunakan agar input bisa berupa bilangan desimal.

Struktur if-elif-else memeriksa kondisi secara berurutan:

- Jika suhu <= 90, maka status Normal.

- Jika tidak memenuhi syarat pertama tapi suhu < 100, maka status Warning.

- Jika tidak memenuhi keduanya (artinya suhu >= 100), maka ALERT: SHUTDOWN.

```python
3. # Data vibrasi dari 5 unit mesin
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]

# Loop untuk memeriksa setiap nilai vibrasi
for i in range(len(vibrasi_mesin)):
    if vibrasi_mesin[i] > 2.0:
        print(f"Warning: Unit {i+1} vibrasi tinggi!")
```

Penjelasan:

range(len(vibrasi_mesin)) menghasilkan indeks dari 0 sampai 4 (untuk 5 unit).

i+1 digunakan karena nomor unit dimulai dari 1 (bukan 0).

Percabangan if vibrasi_mesin[i] > 2.0: digunakan untuk mendeteksi nilai vibrasi tinggi.
