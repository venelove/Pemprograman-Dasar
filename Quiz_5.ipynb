```python
jumlah_pohon = 50
buah_awal = 10
kenaikan = 5

print("Jumlah buah tiap pohon mangga:")
buah = buah_awal

for i in range(1, jumlah_pohon + 1):
    print(f"Pohon ke-{i}: {buah} buah")
    buah += kenaikan

print("\n")

total_buah = 0
i = 1
buah = buah_awal

while i <= jumlah_pohon:
    total_buah += buah
    buah += kenaikan
    i += 1

print("Total semua buah mangga yang dihasilkan dari 50 pohon adalah:", total_buah)
```
