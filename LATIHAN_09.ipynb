```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/orders.csv')
```

```python
df.describe()

rata_rata_payment = df['payment'].mean()
print("Rata-rata payment:", rata_rata_payment)

modus_payment = df['payment'].mode()
print("Modus payment:", modus_payment)
```

```python
import matplotlib.pyplot as plt

df['payment'].hist()
plt.title("Orders")
plt.xlabel("Payment")
plt.ylabel("Customer")
plt.show()
```

```python
corr_matrix = df[['payment']].corr()
print("Korelasi:", corr_matrix)

cov_matrix = df[['payment']].cov()
print("Kovarian:", cov_matrix)
```

```python
df['Kategori_Orders'] = df['payment'].apply(lambda x: 'Banyak' if x >= 3 else 'Sedikit')
df

group_mean = df.groupby('Kategori_Orders')['payment'].mean()
print(group_mean)

group_agg = df.groupby('Kategori_Orders').agg({
    'payment': ['mean', 'max'],
    'payment': ['mean', 'count']
})
group_agg
```
